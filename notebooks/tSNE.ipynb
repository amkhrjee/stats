t-SNE is a dimensionality reduction technique primarily used for visualization of high dimensional data. 

Here, we use it to visualize the MNIST dataset.

```python
import matplotlib.pyplot as plt
from sklearn import datasets

digits = datasets.load_digits()
```

```python
_, axes = plt.subplots(nrows=1, ncols=10, figsize=(10, 10))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r)
```

```python
from sklearn.manifold import TSNE

digits_embedded = TSNE(n_components=2, init="pca", perplexity=50.0).fit_transform(
    digits.data
)
digits_embedded.shape
```

```python
X = digits_embedded
y = digits.target

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c="lightgray", s=5)

for i in range(0, len(X), 5):  # every 5th point
    plt.text(X[i, 0], X[i, 1], str(y[i]), fontsize=8)

plt.xticks([])
plt.yticks([])
plt.show()
```

### PCA

What if we just do PCA?

```python
from sklearn.decomposition import PCA

digits_pca = PCA(n_components=2).fit_transform(digits.data)

X = digits_pca
y = digits.target

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c="lightgray", s=5)

for i in range(0, len(X), 5):  # every 5th point
    plt.text(X[i, 0], X[i, 1], str(y[i]), fontsize=8)

plt.xticks([])
plt.yticks([])
plt.show()
```
