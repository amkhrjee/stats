We will try to fit a linear regression line with gradient descent!

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 100
X = np.random.randn(n)
y = 4 + 3 * X + np.random.randn(n)

plt.scatter(X, y, s=5)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

### Loss function

We will use the mean squared error as our loss function of choice.

$C = \dfrac{1}{n}(y_i - \hat{y_i})^2$ where $\hat{y_i} = mx_i + c$

Therefore, we get $C = \dfrac{1}{n}(y_i - (mx_i + c))^2$

### Gradient descent

We calculate the partial derivative of the loss function with respect to $m$ and $c$.

$C_m = \dfrac{1}{n}\sum_{i=0}^n2(y_i - (mx_i + c))(-x_i) = -\dfrac{2}{n}\sum_{i=0}^nx_i(y_i - (mx_i + c))$ 

$C_c = -\dfrac{2}{n}\sum_{i=0}^n(y_i - (mx_i + c))$

#### Updating $m$ and $c$

$m = m - \eta C_m$

$c = c - \eta C_c$

Here $\eta$ is the learning rate.

```python
# initialization
m = 0
c = 0
eta = 1e-3

record = [(m, c)]

# gradient descent
epochs = 10000
for _ in range(epochs):
    y_hat = m * X + c
    m = m - eta * (-(2 / n) * np.sum(X * (y - y_hat)))
    c = c - eta * (-(2 / n) * np.sum(y - y_hat))
    record.append((m, c))

plt.scatter(X, y, s=5)
plt.xlabel("X")
plt.ylabel("y")
for i, (m, c) in enumerate(record):
    if i % 100 == 0:
        plt.plot(X, m * X + c, color="red", alpha=0.2)
plt.plot(X, m * X + c, color="red")
plt.show()
```
