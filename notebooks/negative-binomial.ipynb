### Negative Binomial Distribution

```python
import random

def bernoulli_process(p=0.5):
    return random.random() < p
```

```python
import matplotlib.pyplot as plt

def count_failures_until_target_success(success_proba=0.3, target_success=20):
    n_success = 0
    n_failures = 0

    while n_success < target_success:
        if bernoulli_process(success_proba):
            n_success += 1
        else:
            n_failures += 1

    return n_failures

success_proba = 0.3
target_success = 400
n_runs = 10000

failures = [
    count_failures_until_target_success(success_proba, target_success)
    for _ in range(n_runs)
]

plt.hist(failures, bins=30, edgecolor="black")
plt.xlabel("Number of failures")
plt.ylabel("Frequency")
plt.title(f"Histogram of failures before {target_success} successes")
plt.show()
```
